##### what is the total revenue for each Brand ?

```python
Brand_by_revenue=df.groupby('brand_name')['revenue'].sum()
Brand_by_revenue
```

```python
plt.bar(Brand_by_revenue.index, Brand_by_revenue.values, width=0.5)
plt.xticks(rotation=45)
plt.show()
```

##### what is the total revenue for each Category Name ?

```python
Category_by_revenue=df.groupby('category_name')['revenue'].sum()
Category_by_revenue
```

```python
plt.bar(Category_by_revenue.index, Category_by_revenue.values, color='green')
plt.xticks(rotation=45)
plt.show()
```

### you should ask 4 questions and answer it

#### top 3 categories sold

```python
category_revenue=df.groupby('category_name')['revenue'].sum()
category_revenue.sort_values(ascending=False).head(3)
```

#### most brand sold

```python
brand_revenue=df.groupby('brand_name')['revenue'].sum()
brand_revenue.nlargest(1)
```

#### What is the most expensive bike?

```python
product_name = df.groupby('product_name')[['revenue', 'total_units']].sum()

# Calculate the price for each product by dividing revenue by total_units
product_name['price'] = product_name['revenue'] / product_name['total_units']
product_name['price'].nlargest(1)
```

####  Which category is most sold?

```python
category_revenue=df.groupby('category_name')['revenue'].sum()
category_revenue.sort_values(ascending=False).head(1)
```
